```r
library(ggplot2)
```

```r
df <- read.csv("timemeasurements1.csv")
df$operation <- as.factor(df$operation)
```

```r
df_avg <- aggregate(time ~ size:operation, data=df, mean)
df_avg$sd <- aggregate(time ~ size:operation, data=df, sd)$time
```

```r
df_avg[df_avg$operation=="dot product",]
```

```r
par(mfrow=c(2, 1))
options(repr.plot.width = 12, repr.plot.height =12)

ggplot(df_avg[df_avg$operation=="dot product",], aes(size, time)) + 
geom_point(color="darkgreen", size=3.5) + 
geom_errorbar(aes(ymin = time-sd, ymax = time+sd, width=2)) +
theme(axis.text = element_text(size=15), axis.title = element_text(size=20), plot.title=element_text(size=25, hjust=0.5)) + 
labs(title="Vector dot product time by size", x="Vector size", y="Time [s]")

ggplot(df_avg[df_avg$operation=="matrix multiplication",], aes(size, time)) + 
geom_point(color="red", size=3.5) + 
geom_errorbar(aes(ymin = time-sd, ymax = time+sd, width=.5)) +
theme(axis.text = element_text(size=15), axis.title = element_text(size=20), plot.title=element_text(size=25, hjust=0.5)) + 
labs(title="Matrix multipilcation time by size", x="Matrix and vector size", y="Time [s]")
```
