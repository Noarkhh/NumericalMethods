```r
library(ggplot2)
library(patchwork)
```

```r
df <- read.csv("timemeasurements1.csv")
df$operation <- as.factor(df$operation)
```

```r
df_avg <- aggregate(time ~ size:operation, data=df, mean)
df_avg$sd <- aggregate(time ~ size:operation, data=df, sd)$time
```

```r
df_avg[df_avg$operation=="dot product",]
```

```r
par(mfrow=c(2, 1))
ggplot(df_avg[df_avg$operation=="dot product",], aes(size, time)) + geom_point()
ggplot(df_avg[df_avg$operation=="matrix multiplication",], aes(size, time)) + geom_point()
# ggarrange(p1, p2)
```
